```python
def links_ads_function():
    
    '''
    Pour récupérer les liens url adspy et facebook pour chaque pub dans la page des résultats de recherche
    
    '''
    tags = driver.find_elements_by_tag_name('a')
    
    href = [elem.get_attribute('href') for elem in tags]
    
    links_ads = []
    links_fb = []
    for i in href:
        if 'https://app.adspy.com/ads/' in i:
            if i not in links_ads:
                links_ads.append(i)
                #print(i)
                
        elif 'https://www.facebook.com/' in i:
        
            if i not in links_fb:
                links_fb.append(i)
                #print(i)

    return links_ads, links_fb

ads, fb = links_ads_function()
time.sleep(2)
```

```python
def product_details(liens_ads):
    
    '''
    Récupérer le nom de chaque produit de la liste des résultats de recherche
    
    '''
    
    likes_nb = []
    comments_nb = []
    shares_nb = []
    country = []
    shop_links = []
    product_name = []
    for y, x in enumerate(liens_ads):
        debut = time.time()
        driver.get(x)
        time.sleep(3)
        
        # Nombre de likes
        likes = driver.find_element_by_class_name('line').text
        
        
        # Nombre de commentaires
        comments = driver.find_element_by_xpath('/html/body/pm-app/div/div[1]/app-ad-details/div/div[2]/div[1]/fb-ad/div/table[2]/tbody/tr/td[2]/div/div[1]/span[2]').text
        
        # Nombre de partages 
        shares = driver.find_element_by_xpath('/html/body/pm-app/div/div[1]/app-ad-details/div/div[2]/div[1]/fb-ad/div/table[2]/tbody/tr/td[2]/div/div[2]/span[2]').text
        
        
        # Pays
        
        country_1 = driver.find_element_by_xpath('/html/body/pm-app/div/div[1]/app-ad-details/div/div[2]/div[1]/fb-ad/div/table[2]/tbody/tr/td[4]/div')
        country_2 = [i.text for i in country_1.find_elements_by_tag_name('span')]
        if country_2 == []:
            country_2 = [country_1.text]
        #for i in country.find_elements_by_tag_name('span'):
         #   print(i.text)
        
        #Shop button
        shop_button = driver.find_element_by_xpath('/html/body/pm-app/div/div[1]/app-ad-details/div/div[2]/div[1]/fb-ad/div/div[2]/table/tbody/tr/td[2]/a')
        shop_link = shop_button.get_attribute('href')
        #driver.get(shop_link)
        #time.sleep(2)
################################################################################################################################        


################################################################################################################################        
        #print(name)
        time.sleep(1)
        #product_name.append(name)
        likes_nb.append(likes)
        comments_nb.append(comments)
        shares_nb.append(shares)
        country.append(country_2)
        shop_links.append(shop_link)
        #time.sleep(1)
        
        
    return likes_nb, comments_nb, shares_nb, country, shop_links

likes_nb, comments_nb, shares_nb, country, shop_links = product_details(ads)
time.sleep(5)
```
